## Présentation des données

```python
image_count = 0
labels = []
train_counts = []
for dirname in os.listdir('C:\\Users\\user\\Downloads\\archive(1)\\split-garbage-dataset\\train'):
    labels.append(dirname)
    image_count = 0
    for img in os.listdir(os.path.join('C:\\Users\\user\\Downloads\\archive(1)\\split-garbage-dataset\\train',dirname)):
        image_count +=1
    train_counts.append(image_count)

print(list(zip(train_counts,labels)))
```

```python
plt.pie(train_counts,
        explode=(0, 0, 0, 0, 0, 0) , 
        labels=labels,
        autopct='%1.1f%%');
plt.title('-- TRAINING SET --');
```

```python
class_weights = []
total_samples = 1768
total_classes = 6
for ele in train_counts:
    result = round(total_samples / (total_classes * ele),2)
    class_weights.append(result)
print(dict(zip(labels,class_weights)))

class_weights = dict(zip(range(6), class_weights))
```
